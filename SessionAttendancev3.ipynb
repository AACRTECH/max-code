### Recorded Sessions

```python
path = r"C:\Users\maxwell.bicking\OneDrive - American Association for Cancer Res\AM24 Attendance\AACR+Annual+Meeting+2024-trackwise/*" #Path can be modified

#xlsx files in the path
file_list = glob.glob(path + "\*.xlsx") #may also need to modify if path is modded

#list of files to merge 
excl_list = []

for file in file_list:
    rec_report = pd.ExcelFile(file) #allows reading multiple tabs
    recorded = pd.read_excel(rec_report, 3) #this is 4th tab, i.e. index 3 
    recorded["Session"] = recorded.iloc[1].values[0] #Session label comes from the A3 cell of each file, which has Session Name
    excl_list.append(recorded.iloc[7:]) #trims the unecessary rows from the top of each file

rec = pd.concat(excl_list, ignore_index= True)
```

```python
rec
```

```python
col_file = pd.read_excel(file_list[0], 3)

rec_cols = col_file.iloc[6].tolist()

rec_cols.append("Session") #Might be a better way to do this but annoyingly had to do to avoid mismatch error

rec_cols
```

```python
rec.columns = rec_cols
```

```python
keycols = ["First Name", "Last Name", "Email", "Session"]
rec['Unique Key'] = rec[keycols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)
```

```python
dedupe_rec = rec.drop_duplicates(subset=['Unique Key'], keep='first')
```

```python
combined_recorded_report = pd.merge(dedupe_rec,
                                    sessiontime,
                                    on=["Session"],
                                    how="outer")
```
